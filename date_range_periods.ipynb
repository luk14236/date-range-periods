```python
import datetime
import pandas as pd
import json

from dateutil.relativedelta import relativedelta
from sqlalchemy import create_engine
```

```python
current_date = datetime.datetime.now().date()
```

```python
periods = [
    ("Current week", current_date - datetime.timedelta(days=current_date.weekday()), current_date),
    ("Current month", current_date.replace(day=1), current_date),
    ("Current quarter", pd.Period(current_date, freq="Q").start_time, current_date),
    ("Current year", current_date.replace(month=1, day=1), current_date),
    ("Last week", current_date - datetime.timedelta(days=current_date.weekday() + 7), current_date - datetime.timedelta(days=current_date.weekday() + 1)),
    ("Last month", current_date.replace(day=1) - datetime.timedelta(days=1), current_date.replace(day=1) - relativedelta(months=+1) - datetime.timedelta(days=1)),
    ("Last quarter", pd.Period(current_date, freq="Q").start_time - datetime.timedelta(days=1), pd.Period(current_date, freq="Q").start_time - relativedelta(months=+1) -  datetime.timedelta(days=1)),
    ("Last year", current_date.replace(month=1, day=1) - datetime.timedelta(days=1), current_date.replace(month=1, day=1)- relativedelta(years=+1) - datetime.timedelta(days=1)),
    ("Last 3 weeks", current_date - datetime.timedelta(days=current_date.weekday() + 7), current_date - datetime.timedelta(days=current_date.weekday())),
    ("Last 6 months", current_date.replace(day=1) - datetime.timedelta(days=1), current_date.replace(day=1) - relativedelta(months=+6) - datetime.timedelta(days=1))
]
```

```python
df = pd.DataFrame(columns=["date_column", "item_type"])
```

```python
for period in periods:
    item_type, start_date, end_date = period
    dates = pd.date_range(start_date, end_date, freq="D")
    df = df.append(pd.DataFrame({"date_column": dates, "item_type": item_type}))
```
